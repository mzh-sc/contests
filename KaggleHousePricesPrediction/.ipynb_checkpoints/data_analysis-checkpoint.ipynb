## Normality

```python
from scipy import stats
from scipy.stats import norm

sns.distplot(df['SalePrice'], fit=norm)

fig = plt.figure()
res = stats.probplot(df['SalePrice'], plot=plt)
```

'SalePrice' is not normal. It shows 'peakedness', positive skewness and does not follow the diagonal line. This is one of the awesome things you can learn in statistical books: in case of **positive skewness, log transformations** usually works well.

See 
* https://machinelearningmastery.com/how-to-transform-data-to-fit-the-normal-distribution/
* Hair et al., 2013, Multivariate Data Analysis, 7th Edition
* https://www.kaggle.com/pmarcelino/comprehensive-data-exploration-with-python#3.-Keep-calm-and-work-smart

**Taking logs means that errors in predicting expensive houses and cheap houses will affect the result equally.**

```python
#applying log transformation
df['SalePrice'] = np.log(df['SalePrice'])
```

```python
#transformed histogram and normal probability plot
sns.distplot(df['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(df['SalePrice'], plot=plt)
```

```python
#histogram and normal probability plot
sns.distplot(df['GrLivArea'], fit=norm);
fig = plt.figure()
res = stats.probplot(df['GrLivArea'], plot=plt)
fig = plt.figure()
plt.scatter(df['GrLivArea'], df['SalePrice']);
```

# Train model

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 1. don't use scaling, dummy encoding
# performance RMSE: 0.17229534248006997
# Witout log (initial) 38788.796429879745
# With log (initial) 49086.81302470668
# With engeneering of grage and bsmt 36160.34821735394
model_features = list(features)
model_features.remove("SalePrice")
model_features
```

```python
df_final_copy = df_final.copy()
```

```python
# 2. use dummy encoding
# 35421.9977135352
df_final_copy = df_final.copy()



ordinal_columns = ['OverallQual']
for col in ordinal_columns:
    df_final_copy[col] = df_final_copy[col].astype('category')
        
df_final_copy = pd.concat([df_final_copy, pd.get_dummies(df_final_copy[ordinal_columns])], axis=1) \
    .drop(ordinal_columns, axis=1)
    
df_final_copy.head()

model_features = df_final_copy.columns.tolist()
model_features.remove('SalePrice')
model_features
```

```python
# 2.1. scaling
# has no effect on the result since LinearRegression taking care of it by default
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_final_copy[model_features] = scaler.fit_transform(df_final_copy[model_features])
```

```python
lr = LinearRegression()

kf = KFold(n_splits=5, shuffle=True, random_state=1)

rmse_values = []
for train_index, test_index in kf.split(df_final):
    
    train = df_final_copy.iloc[train_index]
    test = df_final_copy.iloc[test_index]
    
    lr.fit(train[model_features], train["SalePrice"])
    
    predictions = lr.predict(test[model_features])
    
    # use np.exp() to return to the original SalePrice scale
    mse = mean_squared_error(np.exp(test["SalePrice"]), np.exp(predictions))
    rmse = np.sqrt(mse)
    rmse_values.append(rmse)
    
print(f'RMSE: {np.mean(rmse_values)}')
```
